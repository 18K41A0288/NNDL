```python
import math
import random as r
x=[0.2,0.4,0.6,0.8]
y=[3.4,3.8,4.2,4.6]
l=0.02
m=r.randint(-50,50)
c=r.randint(-50,50)
print("m = ",m,"c = ",c)
m_iter=int(input("enter maximum number of iterations  "))


def gm(x,y,m,c):
  s = 0
  ns = len(x)
  for i in range(ns):
    gm_i = (y[i]- m*x[i] - c)*x[i]
    s = s + gm_i
  grm = -1*(1/ns)*s
  return grm

def gc(x,y,m,c):
  s = 0
  ns = len(x)
  for i in range(ns):
    gc_i = (y[i]- m*x[i] - c)
    s = s + gc_i
  grc = -1*(1/ns)*s
  return grc

for i in range(m_iter):
  dm = -1*l*gm(x,y,m,c)
  dc = -1*l*gc(x,y,m,c)
  m = m + dm
  c = c + dc
  print("m = ",m,"\t c = ",c)

print("optimal mean value occurs at m = ",m,"c = ",c)
```
