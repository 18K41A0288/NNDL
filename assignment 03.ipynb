```python
import random as r
x=[0.2,0.4,0.6,0.8]
y=[3.4,3.8,4.2,4.6]
l=0.02
m=r.randint(-50,50)
c=r.randint(-50,50)
print("m = ",m,"c = ",c)
m_iter=int(input("enter maximum number of iterations  "))
gm = lambda x,y,m,c: -1*(y-m*x-c)*x
gc = lambda x,y,m,c: -1*(y-m*x-c)

for i in range(m_iter):
  for s in range(4):
    dm = -1*l*gm(y[s],x[s],m,c)
    dc = -1*l*gc(y[s],x[s],m,c)
    m=m+dm
    c=c+dc
    print("m = ",m,"\t c = ",c)
    s+=1

print("optimal mean value occurs at m = ",m,"c = ",c)
```
