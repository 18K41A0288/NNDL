```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split
import math
import random as r
from random import shuffle as rs


data = pd.read_csv("/content/drive/MyDrive/kc_house_data.csv")
l=0.02
m=r.randint(-50,50)
c=r.randint(-50,50)
print("m = ",m,"c = ",c)
m_iter=int(input("enter maximum number of iterations  "))


x = data['sqft_living']
y = data['price']

x = minmax_scale(x)
y = minmax_scale(y)

cd = [(x[i],y[i]) for i in range(len(x))]
cd_train, cd = train_test_split(cd,test_size=0.1)

cd_train, cd_test = train_test_split(cd,test_size=0.1)
print(len(cd_train))
print(len(cd_train))
print(len(cd_test))
print(len(cd_test))

bs=int(input("enter batch size "))
ns=len(cd_train)
nb=ns//bs


def gm(d,m,c):
  s = 0
  for i in range(len(d)):
    s = s + (d[i][1]-m*d[i][0]-c)*d[i][0]
  return -1*s/len(d)

def gc(d,m,c):
  s = 0
  for i in range(len(d)):
    s = s + (d[i][1]-m*d[i][0]-c)
  return -1*s/len(d)

for i in range(m_iter):
    for j in range(nb+1):
      if j == nb:
        d = cd_train[j*bs:]
      else:
        d = cd_train[j*bs:(j+1)*bs]
      dm = -1*(l*gm(d,m,c))
      dc = -1*(l*gc(d,m,c))
      m=m+dm
      c=c+dc
      print("m = ",m,"\t c = ",c)
    r.shuffle(d)
print("\n\noptimal mean value occurs at m = ",m,"c = ",c)
        
```
