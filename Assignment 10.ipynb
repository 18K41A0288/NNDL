```python
import pandas as pd
import numpy as np
import math
import random as r
from random import shuffle
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split

l=0.02
vm=vc=0
mc=0.9
m=r.randint(-5,5)
c=r.randint(-5,5)
print("m = ",m,"c = ",c)
m_iter=int(input("enter maximum number of iterations  "))
bs=int(input("enter batch size : "))


di = pd.read_excel("/content/drive/MyDrive/Load Data in kW.xlsx")

ns=int(len(di)/bs)

x = []
y = []

for i in range(len(di.iloc[:,0])-24):
  x.append(di.iloc[i,0])
  y.append(di.iloc[i+24,0])


def err(d,m,c):
  s = 0
  for i in range(len(d)):
    s = s + math.pow(d[i][1]-m*d[i][0]-c,2)
  return 0.5*s/len(d)

def gm(d,m,c):
  s = 0
  for i in range(len(d)):
    s = s + (d[i][1]-m*d[i][0]-c)*d[i][0]
  return -1*s/len(d)

def gc(d,m,c):
  s = 0
  for i in range(len(d)):
    s = s + (d[i][1]-m*d[i][0]-c)
  return -1*s/len(d)

max = di.min()
min = di.max()

data=[]
for i in range(ns):
  data.append((x[i],y[i]))

data_train,data_test = train_test_split(data,test_size=0.1)

error=[]
nb = (len(data_train)//bs)
for i in range(m_iter):
  for j in range(nb+1):
    if j == nb:
      d1 = data_train[j*bs:]
    else:
      d1 = data_train[j*bs:(j+1)*bs]
    
    dm = -1*l*gm(d1,m,c)
    dc = -1*l*gc(d1,m,c)

    vm = vm*mc + dm
    vc = vc*mc + dc
    
    m = m + vm
    c = c + vc

    print("epoch {}, batch {}, error {}".format(i,j,err(d1,m,c)))
    error.append(err(d1,m,c))
  r.shuffle(data_train)
m=1.183121
c=6.000093
mn=3377.9196
mx=805444.4304

iload = mn + (mx-mn)*r.random()
print("input load is =",iload)

print("m = ",m,"c = ",c,)
4
output = iload*m+c
print("predicted load for this hour is : ",output)
```
