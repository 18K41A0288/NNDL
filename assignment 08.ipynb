```python
import math
import random as r
import pandas as pd
import numpy as np
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split

x=[0.2,0.4,0.6,0.8]
y=[3.4,3.8,4.2,4.6]
l=0.02
m=r.randint(-50,50)
c=r.randint(-50,50)
print("m = ",m,"c = ",c)
m_iter=int(input("enter maximum number of iterations  "))

x=[]
y=[]
data = pd.read_excel("/content/drive/MyDrive/slr08.xls")

y = data.iloc[:,1]
x = data.iloc[:,0]
x = minmax_scale(x)
y = minmax_scale(y)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

x2 = x_test
y2 = [m*i+c for i in x2]

x1 = x_test
y1 = y_test

def gm(x,y,m,c):
  s = 0
  ns = len(x)
  for i in range(ns):
    gm_i = (y[i]- m*x[i] - c)*x[i]
    s = s + gm_i
  grm = -1*(1/ns)*s
  return grm

def gc(x,y,m,c):
  s = 0
  ns = len(x)
  for i in range(ns):
    gc_i = (y[i]- m*x[i] - c)
    s = s + gc_i
  grc = -1*(1/ns)*s
  return grc

def er(x,y,m,c):
  ns = len(x)
  for i in range(ns):
    error = 1*0.5*(1/ns)*math.pow((y[i] - m*x[i] - c), 2)
  return error

for i in range(m_iter):
  dm = -1*l*gm(x,y,m,c)
  dc = -1*l*gc(x,y,m,c)
  m = m + dm
  c = c + dc
  print("m = ",m,"\t c = ",c)


train_error = er(x_train,y_train,m,c)
test_error = er(x_test,y_test,m,c)

print("training error : ",train_error)
print("testing error : ",test_error)
pressure_max = 120.000000
pressure_min = 62.200000

weight_max = 578.600000
weight_min = 406.700000

actual_pressure = float(input("enter the pressure value : "))
scaled_pressure = (actual_pressure - pressure_min)/(pressure_max-pressure_min)

scaled_weight = scaled_pressure*m+c
predicted_weight = scaled_weight*(weight_max-weight_min)+weight_min

print("the predicted weight : ",predicted_weight)

print("optimal mean value occurs at m = ",m,"c = ",c)
```
