find global minimum point and value for function f(x,y)= x^4+y^2+10

```
import random as r
x = r.randint(-10,10)
y = r.randint(-10,10)
lr =r.random()
print("x = ",x,"\ty= ",y,"\tlr = ",lr)
mi = int(input('enter maximum iterations - '))
ps = 1 
i=0
f = lambda x,y: (x**2)+(y**2)+10
dfx = lambda x: (2*(x))
dfy = lambda y: (2*(y))
while i < mi:
    xi =x
    yi=y 
    print('f(x,y)= ',f(xi,yi))
    x = x - lr * dfx(xi)
    y = y - lr * dfy(yi)
    ps = abs(x -xi) 
    ps = abs(y -yi)
    i+=1
    print("Iteration number = ",i,"\nX value is",x,"\nY value is",y)    
print("minimum value of F(X) = ",f(xi,yi),"occurs at x= ", x," and y =",y)
```
