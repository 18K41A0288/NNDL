```python
import pandas as pd
import numpy as np
import math
import random as r
import matplotlib.pyplot as plt
from random import shuffle
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split

load_data = pd.read_excel("/content/drive/MyDrive/Load Data in kW.xlsx")

l=0.02
vm=vc=0
mc=0.9
m=r.randint(-5,5)
c=r.randint(-5,5)
print("m = ",m,"c = ",c)
m_iter=int(input("enter maximum number of iterations  "))
bs=int(input("enter batch size : "))


def error(d,m,c):
  s = 0
  for i in range(len(d)):
    s = s + math.pow(d[i][1]-m*d[i][0]-c,2)
  return 0.5*s/len(d)

def gm(d,m,c,vm,vc,mc):
  s = 0
  for i in range(len(d)):
    s = s + (d[i][1]-(m+mc*vm)*d[i][0]-(c+mc*vc))*d[i][0]
  return -1*s/len(d)

def gc(d,m,c,vm,vc,mc):
  s = 0
  for i in range(len(d)):
    s = s + (d[i][1]-(m+mc*vm)*d[i][0]-(c+mc*vc))
  return -1*s/len(d)


x = []
y = []
for i in range(len(load_data.iloc[:,0])-24):
  x.append(load_data.iloc[i,0])
  y.append(load_data.iloc[i+24,0])

x = minmax_scale(x)
y = minmax_scale(y)

d = []

for i in range(len(x)):
  d.append((x[i],y[i]))

print(len(d))

d_train,d_test = train_test_split(d,test_size=0.1)

len(d_train)

err= []
nb = (len(d_train)//bs)
for i in range(m_iter):
  for j in range(nb+1):
    if j == nb:
      d = d_train[j*bs:]
    else:
      d = d_train[j*bs:(j+1)*bs]

    dm = -1*l*gm(d,m,c,vm,vc,mc)
    dc = -1*l*gc(d,m,c,vm,vc,mc)

    vm = vm*mc + dm
    vc = vc*mc + dc

    m = m + vm
    c = c + vc

    print("epoch {}, batch {}, error {}".format(i,j,error(d,m,c)))
    err.append(error(d,m,c))
  r.shuffle(d_train)

  
input_load = float(input("enter yesterday load at this time : "))

max_value = 8841.669480
min_value = 3377.919600

input_load = (input_load - min_value)/(max_value-min_value)
output_load = input_load*m+c

output_load = output_load*(max_value-min_value)+min_value

print("predicted load for this hour is : ",output_load)
```
