```python
import random as r
x=[0.2,0.4,0.6,0.8]
y=[3.4,3.8,4.2,4.6]
l=0.02
m=r.randint(-50,50)
c=r.randint(-50,50)
print("m = ",m,"c = ",c)
m_iter=int(input("enter maximum number of iterations  "))
ns=len(x)
bs=int(input("enter batch size : "))
nb=int(ns/bs)
b=list()
for i in range(ns):
  b.append((x[i],y[i]))

def gm(d,m,c):
  s = 0
  for i in range(len(d)):
    s = s + (d[i][1]-m*d[i][0]-c)*d[i][0]
  return -1*s/len(d)

def gc(d,m,c):
  s = 0
  for i in range(len(d)):
    s = s + (d[i][1]-m*d[i][0]-c)
  return -1*s/len(d)

for i in range(m_iter):
  for j in range(nb):
    if j == nb:
      d = b[j*bs:]
    else:
      d = b[j*bs:(j+1)*bs]
    dm = -1*(l*gm(d,m,c))
    dc = -1*(l*gc(d,m,c))
    m=m+dm
    c=c+dc
    print("m = ",m,"\t c = ",c)
  r.shuffle(d)
print("\n\noptimal mean value occurs at m = ",m,"c = ",c)
```
