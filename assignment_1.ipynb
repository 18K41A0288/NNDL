```
x = int(input('enter initial value - '))
lr = float(input('enter Learning rate - '))
a =float(input('enter accuracy - '))
mi = int(input('enter maximum iterations - '))
ps = 1 
i=0
f = lambda x: (x**4)+(3*x**2)+10
df = lambda x: (4*(x**3))+(6*x)
while ps > a and i < mi:
    xi =x 
    print('f(x)= ',f(xi))
    x = x - lr * df(xi)
    ps = abs(x -xi) 
    i+=1
    print("Iteration number = ",i,"\nX value is",x)    
print("minimum value of F(X) = ",f(xi),"occurs at", round(x))
```
